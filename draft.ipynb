```python
import numpy as np
import pandas as pd
```

```python
sentence = 'Taclḥiyt nɣ Taclḥit nɣ Tasusit tga yat tutlayt tamaziɣt illan ɣ lmɣrib, middn li sis isawaln gan Iclḥiyn.'
token_sequence = str.split(sentence)
vocab = sorted(set(token_sequence))
# ', '.join(vocab)
num_tokens = len(token_sequence)
vocab_size = len(vocab)
onehot_vectors = np.zeros((
    num_tokens,
    vocab_size), int)
for i, word in enumerate(token_sequence):
    onehot_vectors[i, vocab.index(word)] = 1
```

```python
df = pd.DataFrame(onehot_vectors, columns=vocab)
df[df==0]=' '
df
```

```python
sentence_bow = {}
for token in sentence.split():
    sentence_bow[token] = 1
sorted(sentence_bow.items())
```

```python
df = pd.DataFrame(pd.Series(dict([(token, 1) for token in sentence.split()])), columns=['sent']).T
df
```

```python
sentences = """Tettusawal Tclḥit ɣ Sus, Iḥaḥan, Lḥuz n Mṛṛakc d kra n tsggʷin yaḍnin asn d iwalan.\n"""
sentences += """Tumẓ tutlayt taclḥit zɣ Taṣṣurt ɣ ugafa ar Agʷlmim ɣ unẓul ar Warzazat ɣ ugmuḍ.\n"""
sentences += """Mac ar tt inn nttafa tlla ula ɣ timdinin mqqurnin n Lmɣrib zun d Tigmmi Tumlilt (kaẓa), d Ṛṛbaḍ d Sla.\n"""
sentences += """Tkka tt inn tclḥiyt ar tettyura s uskkil aɛrab, ilin sis kigan n warratn iclḥiyn."""
corpus = {}
for i, sent in enumerate(sentences.split('\n')):
    corpus['sent{}'.format(i)] = dict((tok, 1) for tok in sent.split())

df = pd.DataFrame.from_records(corpus).fillna(0).astype(int).T
df[df.columns[:10]]
```

```python
# The plus sign after the closing
# square bracket ( ] ) means that a match must contain one or more of the characters
# inside the square brackets.

import re
sentence = """Ar ass-ad ur illi kra n uṣiḍn igaddan ittuyskarn f imsiwln n tutlayin timaziɣin ɣ tmazirt n Lmɣrib."""
tokens = re.split(r'[-\s.,;!?]+', sentence)
tokens
```

```python
pattern = re.compile(r'[-\s.,;!?]+')
tokens = pattern.split(sentence)
tokens
```

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+|$[0-9.]+|\S+-')
tokenizer.tokenize(sentence)
```

```python
# Treebank Tokenizer
from nltk.tokenize import TreebankWordTokenizer
tokenizer = TreebankWordTokenizer()
tokenizer.tokenize(sentence)
```
